```python
fact=spark.table('databricks_catalog.gold.fact_trip')
```

#####No duplicate trip_key

```python
dup_cnt=fact.groupBy('trip_key').count().filter('count>1').count()

if dup_cnt > 0:
    raise Exception("DQ FAILED: duplicate trip_key found")
```

#####No NULL surrogate keys

```python
null_fk_cnt = fact.filter(
    fact.pu_location_sk.isNull() |
    fact.do_location_sk.isNull() |
    fact.payment_sk.isNull() |
    fact.ratecode_sk.isNull()
).count()

if null_fk_cnt > 0:
    raise Exception("DQ FAILED: NULL foreign keys found")
```

####No Negative values

```python
neg_cnt = fact.filter(
    (fact.trip_distance < 0) |
    (fact.fare_amount < 0) |
    (fact.total_amount < 0)
).count()

if neg_cnt > 0:
    raise Exception("DQ FAILED: negative values found")
```
