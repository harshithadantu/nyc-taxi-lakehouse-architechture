####Trips per day

```python
%sql
SELECT
  trip_date,
  COUNT(*) AS trips
FROM databricks_catalog.gold.fact_trip
GROUP BY trip_date
ORDER BY trip_date;
```

####Total revenue per day

```python
%sql
SELECT
  trip_date,
  ROUND(SUM(total_amount), 2) AS total_revenue
FROM databricks_catalog.gold.fact_trip
GROUP BY trip_date
ORDER BY trip_date;
```

#####Top pickup zones (historic, correct)

```python
%sql
SELECT
  l.borough,
  l.zone,
  COUNT(*) AS trips
FROM databricks_catalog.gold.fact_trip f
JOIN databricks_catalog.gold.dim_location_scd2 l
  ON f.pu_location_sk = l.location_sk
 AND f.pickup_ts >= l.effective_from
 AND f.pickup_ts <  COALESCE(l.effective_to, TIMESTAMP '2999-12-31')
GROUP BY l.borough, l.zone
ORDER BY trips DESC
LIMIT 10;
```
