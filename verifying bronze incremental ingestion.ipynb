```python
from pyspark.sql import functions as F

data = [
    (1, "2021-04-01 08:15:00", "2021-04-01 08:30:00", 132, 170, 1, 2.1, 9.5, 1, 12.0),
    (2, "2021-04-02 14:30:00", "2021-04-02 14:50:00", 48, 68, 2, 3.5, 14.0, 2, 17.5),
    (1, "2021-04-03 19:45:00", "2021-04-03 20:05:00", 138, 107, 1, 1.8, 8.0, 1, 10.5),
]

columns = [
    "VendorID",
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime",
    "PULocationID",
    "DOLocationID",
    "passenger_count",
    "trip_distance",
    "fare_amount",
    "payment_type",
    "total_amount",
]

df = (
    spark.createDataFrame(data, columns)
    .withColumn("tpep_pickup_datetime", F.to_timestamp("tpep_pickup_datetime"))
    .withColumn("tpep_dropoff_datetime", F.to_timestamp("tpep_dropoff_datetime"))
)
```

```python
df.coalesce(1).write.mode("overwrite").parquet(
    "abfss://source@databrickseteprojectsa.dfs.core.windows.net/yellow_tripdata/yellow_tripdata_2021-04.parquet"
)
```
